Intro & Problem:_
The objective of this notebook is to understand how to read, summarize, and visualize a dataset. By utilizing pandas, matplotlib, and seaborn, I will perform Exploratory Data Analysis (EDA) to find patterns within the data.

Dataset Understanding and Description:_
For this task, I am using the standard Iris Dataset.

1-Structure: After loading the data, I determined that it contains 150 rows and 6 columns. 
2-Features: The data points being analyzed include a list of your column names here, like Sepal Length and Sepal Width.
3-Goal: The ultimate goal is to see how these physical measurements relate to the different Iris flower species.

```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sb
```

```python
irisfile=pd.read_csv("Iris.csv")
```

```python
irisfile.columns
```

```python
irisfile.head()
```

```python
sb.scatterplot(x="SepalLengthCm",y="SepalWidthCm",data=irisfile)
mp.show()
```

```python
sb.histplot(x="SepalLengthCm",y="SepalWidthCm",data=irisfile)
mp.show()
```

```python
sb.boxplot(x="SepalLengthCm",y="SepalWidthCm",data=irisfile)
mp.show()
```

Conclusion and Key Insights
After generating the visual data, here are the primary takeaways from the Iris dataset:
Scatter Plot Insights:By mapping X column against Y column, I observed that dots clustered together by species and were not displaying true message.
Histogram Insights: The distribution of the data shows that most of the measurements fall between SepalWidth of 4.5 and SepalLength 5.5.
Box Plot Insights: When analyzing the data for outliers I saw dots outside the boxes stating that the data need further cleaning and manipulation.
